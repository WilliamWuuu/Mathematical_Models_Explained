## 时间序列分析中常用的三种模型：
- 季节分解模型
- 指数平滑方法
- ARIMA模型
## 时间序列分解
数据具有周期性时才能使用时间序列分解，例如数据是月份数据(周期为12)、季度数据(周期为4)，如果是年份数据则不行。
- 长期变动趋势(Secular Trend, T): 在相当长的一段时间内，受到长期趋势影响因素的影响，表现出持续上升或持续下降的趋势。
- 季节变动(Seasonal Variation, S): 由于季节的转变使得指标数值发生周期性变动。这里的季节是广义的，一般以月、季、周为时间单位，不能以年作单位。
- 周期变动(Cyclical Variation, C): 通常以若干年为周期, 增加和减少交替出现，但是并不具严格规则的周期性连续变动, 曲线图上表现为波浪式的周期变动。
- 不规则变动(随机扰动项)(Irregular Variation, I): 某些随机因素导致的数值变化，这些因素的作用是不可预知和没有规律性的
#### 四种变动与指标数值最终变动(Y)的关系可能是叠加关系，也可能是成绩关系
- 如果四种变动之间是相互独立的关系，那么叠加模型可以表示为：
$$Y=T+S+C+I$$
- 如果随着时间的推移，时间序列图的波动保持恒定，则可以直接使用叠加模型
- 如果四种变动之间存在相互影响关系，那么应该使用乘积模型
$$Y=T×S×C×I$$
- 如果随着时间的推移，序列的季节波动变得越来越大，则反映各种变动之间的关系发生变化，建议使用乘积模型
- 如果不存在季节波动，则两种分解均可以
